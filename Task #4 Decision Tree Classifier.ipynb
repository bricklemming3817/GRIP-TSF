# Task #4: Decision Tree Classifier
In this task we will create a Decision Tree Classifier using Iris dataset and visualise it.

### 1. Importing Libraries

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets as datasets
```

### 2. Loading Dataset

```python
iris = datasets.load_iris()
data = pd.DataFrame(iris.data, columns = iris.feature_names)
y = iris.target
```

### 3. Initialising Algorothm
Here we will be initialising the Decision Tree Classifier and dataset is fitted using the dataset and clusters.

```python
dtree = DecisionTreeClassifier()
dtree.fit(data, y)
```

### 4. Visualising Decision Tree

```python
fn=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
cn=['setosa', 'versicolor', 'virginica']
fig, axs = plt.subplots(figsize=(20,20))
tree.plot_tree(dtree,feature_names = fn, 
               class_names=cn,
               filled = True)
plt.title("Decision Tree")
plt.show()
```
